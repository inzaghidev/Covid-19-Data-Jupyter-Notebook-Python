# Data COVID-19 di Seluruh Dunia

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

## Jumlah Kasus COVID-19 Seluruh Dunia

```python
cases_data_worldwide = pd.read_csv('./csv/covid-19_data_cases_worldwide.csv')
```

```python
dates_every_90_days_greg = cases_data_worldwide['tanggal_masehi'][::90]
dates_every_90_days_hijri = cases_data_worldwide['tanggal_hijriah'][::90]
```

```python
cases_data_worldwide.head()
```

```python
cases_data_worldwide.head(50)
```

```python
# Define the function to convert numbers to millions format
def millions_format(x, pos):
    'The two args are the value and tick position'
    return '{:1.1f}M'.format(x * 1e-6)

# Step 2: Create the plot
plt.figure(figsize=(10, 6))
plt.plot(cases_data_worldwide['tanggal_masehi'], cases_data_worldwide['confirmed_cases'], label='Confirmed Cases', color='yellow')
plt.plot(cases_data_worldwide['tanggal_masehi'], cases_data_worldwide['deaths'], label='Death Cases', color='red')
plt.plot(cases_data_worldwide['tanggal_masehi'], cases_data_worldwide['recoveries'], label='Recoveried Cases', color='green')
plt.plot(cases_data_worldwide['tanggal_masehi'], cases_data_worldwide['active_cases'], label='Active Cases', color='blue')
plt.plot(cases_data_worldwide['tanggal_masehi'], cases_data_worldwide['passive_cases'], label='Passive Cases', color='orange')

plt.xlabel('Date')
plt.ylabel('Jumlah Kasus (Juta)')
plt.title('Kurva Kasus Konfirmasi COVID-19 di Seluruh Dunia (Tanggal Masehi)')
plt.legend()
plt.xticks(dates_every_90_days_greg, rotation=45)  # Set the x-axis ticks to the dates every 100 days

# Format the y-axis labels in millions format
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(millions_format))

plt.show()
```
